```python
import numpy as np
x = ["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]
p = np.arange(len(x))
width = 0.2
p1 = [j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Popularity No.",fontsize=15)
plt.title("Coding Languages")
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
